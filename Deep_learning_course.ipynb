### Subplots

plt.subplot(a,b,c)
Means:
a -> number of rows
b -> number of columns
c -> subplot index

eg:
plt.subplot(1,2,1)
1 row
2 column
1 subplot index

```python
import matplotlib.pyplot as plt
import numpy as np

plt.subplot(2,2,1)
plt.plot(np.random.randn(10))

plt.subplot(2,2,4)
plt.plot(np.random.randn(10))

```

plt.subplots(f,a,figsize)

f - figure
a - axis
figsize (actual name) -> optional parameter size of figure

eg: plt.subplots(1,2)


OR

plt.subplots(row,column)


```python
import numpy as np
import matplotlib.pyplot as plt

fig,ax = plt.subplots(1,2,figsize=(10,4))

x = np.arange(10)
ax[0].plot(x,x**2,'b')
ax[0].plot(x,x,'k')

ax[1].plot(x,np.sqrt(x),'r')

```

```python
import numpy as np
import matplotlib.pyplot as plt

fig,ax = plt.subplots(2,2)
ax[0,0].plot(np.random.randn(4,4))
ax[0,1].plot(np.random.randn(4,4))
ax[1,0].plot(np.random.randn(4,4))
ax[1,1].plot(np.random.randn(4,4))

plt.tight_layout()
plt.show()
```

### Seaborn library

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import seaborn as sns

n = 200
D = np.zeros((n,2))

D[:,0] = np.linspace(0,10,n) + np.random.randn(n)
D[:,1] = D[:,0]**2 + np.random.randn(n) * 10

# plt.plot(D[:,0],D[:,1],'o')
sns.jointplot(x = D[:,0], y = D[:,1])
```

```python
df = pd.DataFrame(data=D, columns=['var1','var2'])
sns.jointplot(x = df.columns[0], y = df.columns[1], data=df, kind = 'kde', color = 'purple')
```

```python
x = np.linspace(-1,1,n)
y1 = x ** 2
y2 = np.sin(3*x)
y3 = np.exp(-10*x**2)

sns.scatterplot(x=y1,y=y2,hue=y3)
```

### Images

```python
# we can plot a matrix but images gives better visuals

import numpy as np
import matplotlib.pyplot as plt

row = 3
col = 5
M = np.random.randint(10, size=(row,col))
print(M)
# plt.plot(M)

plt.imshow(M)

for i in range(row):
  for j in range(col):
    plt.text(j,i,str(M[i,j]), fontsize = 15,
             horizontalalignment='center',
             verticalalignment='center')

plt.colorbar()
plt.show()
```

```python
# EXCERCISE: Visualize the hilbert matrix using seaborn (heatmap)
# H is 10x10

import seaborn as sns

n = 10
H = np.zeros((n,n))

for i in range(n):
  for j in range(n):
    H[i,j] = 1 / (((i+1) + (j+1)) - 1)

sns.heatmap(H)
plt.show()
```
